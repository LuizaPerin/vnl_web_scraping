```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
df = pd.DataFrame()

years = ['2021', '2022', '2023', '2024']
categories = ['best-attackers', 'best-blockers', 'best-servers', 'best-setters', 'best-diggers', 'best-receivers']

def fetch_data(year, category):
    url = f'https://en.volleyballworld.com/volleyball/competitions/volleyball-nations-league/{year}/statistics/men/{category}/'
    page = requests.get(url)
    page.encoding = 'utf-8'
    soup = BeautifulSoup(page.text, 'html')

    table = soup.find_all('table')[0]
    titles = [title.text for title in table.find_all('span', class_='full-text')]

    data = []
    for row in table.find_all('tr')[1:]:
        row_data = [data.text.strip() for data in row.find_all('td')]
        data.append(row_data)

    df = pd.DataFrame(data, columns=titles)
    df['Year'] = year
    return df

all_data = {}

for category in categories:
    category_dfs = []
    for year in years:
        df = fetch_data(year, category)
        category_dfs.append(df)
    all_data[category] = pd.concat(category_dfs, ignore_index=True)

# dataframe por categoria
best_attackers_df = all_data['best-attackers']
best_blockers_df = all_data['best-blockers']
best_servers_df = all_data['best-servers']
best_setters_df = all_data['best-setters']
best_diggers_df = all_data['best-diggers']
best_receivers_df = all_data['best-receivers']
```

```python
from google.colab import files

def save_and_download_df(df, filename):
    df.to_csv(filename, encoding='utf-8-sig', index=False)
    files.download(filename)

dataframes = {
    'best_attackers_df': best_attackers_df,
    'best_blockers_df': best_blockers_df,
    'best_servers_df': best_servers_df,
    'best_setters_df': best_setters_df,
    'best_diggers_df': best_diggers_df,
    'best_receivers_df': best_receivers_df
}

# loop para salvar e baixar cada DataFrame
for name, df in dataframes.items():
    filename = f'men_{name}.csv'
    save_and_download_df(df, filename)
```
