```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
df = pd.DataFrame()

years = ['2021', '2022', '2023', '2024']

for year in years:
    url = f'https://en.volleyballworld.com/volleyball/competitions/volleyball-nations-league/{year}/statistics/men/best-scorers/'

    page = requests.get(url)
    page.encoding = 'utf-8'

    soup = BeautifulSoup(page.text, 'html')

    table = soup.find_all('table')[0]
    titles = table.find_all('span', class_ = 'full-text')
    table_titles = [title.text for title in titles]

    df_year = pd.DataFrame(columns = table_titles)

    column_data = table.find_all('tr')

    for row in column_data[1:]:
      row_data = row.find_all('td')
      individual_row_data = [data.text for data in row_data]
      # print(individual_row_data)

      lenght = len(df_year)
      df_year.loc[lenght] = individual_row_data

    df_year['Year'] = year

    df = pd.concat([df, df_year])

df
```

```python
from google.colab import files
df.to_csv('men_best_scorers.csv', encoding = 'utf-8-sig')
files.download('men_best_scorers.csv')
```
